# K Nearest Neighbors with Python

We have a classified data set where the feature column names are anonymized but the target column is known. 

We'll try to use KNN to create a model that directly predicts a class for a new data point based off of the features.

## Import Libraries



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Get the Data

Set index_col=0 to use the first column as the index.

```python
df = pd.read_csv("./../../data/classified_data.csv",index_col=0)
```

```python
df.head()
```

## Standardize the Variables

Because the KNN classifier predicts the class of a given test observation by identifying the observations that are nearest to it, the scale of the variables matters.

Any variables that are on a large scale will have a much larger effect on the distance between the observations, and hence on the KNN classifier, than variables that are on a small scale.

Example:
Say you have two features:

age ranging from 18–70

income ranging from 20,000–200,000

Without scaling, income dominates the distance calculation just because it's a larger number — even if it's not more important. That leads to misleading neighbors and poor model performance.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

**We only need to standardize the scale for features, not the target class**

```python
scaler.fit(df.drop('TARGET CLASS',axis=1))
scaled_features = scaler.transform(df.drop('TARGET CLASS',axis=1))
df_feat = pd.DataFrame(scaled_features,columns=df.columns[:-1])
# Check the head of the scaled data. Not going to see a visible difference due to randamness of the data, but the values are now standardized regardless.
df_feat.head()
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled_features,df['TARGET CLASS'],
                                                    test_size=0.30)
```

## Using KNN

Remember that we are trying to come up with a model to predict whether someone will TARGET CLASS or not. We'll start with k=1.

**n_neighbors** is the number of neighbors the algorithm considers when making a prediction

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
```

## Predictions and Evaluations

Let's get the predictions

```python
pred = knn.predict(X_test)
```

Let's evaluate our KNN model!

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

## Choosing a K Value

Let's go ahead and use the elbow method to pick a good K Value.

For this we'll iterate through a range for candidate K values and collect the resulting predictions.

Then, we'll compare these predictions with the corresponding error rate

```python
error_rate = []
for i in range(1,40):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    #pred_i != y_test: This returns a Boolean array where each value is:
    # True if the prediction is incorrect & False if the prediction is correct
    # np.mean(pred_i != y_test): This returns the mean of the Boolean array, which is the error rate
    # For example: [False, True, False] → [0, 1, 0] → np.mean = 1/3 = 0.333... i.e, the model got 1 out of 3 predictions wrong → 33.3% error rate.
    error_rate.append(np.mean(pred_i != y_test))
```

**Let's plot a chart for 'Error Rate' vs 'K'**

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

Here we can see that that after around K>15 the error rate just tends to hover around 0.06-0.05 Let's retrain the model with that and check the classification report!

```python
# FIRST A QUICK COMPARISON TO OUR ORIGINAL K=1
knn = KNeighborsClassifier(n_neighbors=1)

knn.fit(X_train,y_train)
pred = knn.predict(X_test)

print('WITH K=1')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```

```python
# NOW WITH K=23
knn = KNeighborsClassifier(n_neighbors=23)

knn.fit(X_train,y_train)
pred = knn.predict(X_test)

print('WITH K=15')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```

Our model performs a bit better after tuning to a better K value
